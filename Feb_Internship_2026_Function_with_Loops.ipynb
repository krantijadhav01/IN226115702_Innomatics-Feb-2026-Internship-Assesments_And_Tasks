```python
#Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
  valid_plans = [199, 299, 399, 599]

  if amount < 50:
    print("Recharge amount must be tleast 50")
    return False

  if amount not in valid_plans:
    print("Invalid recharge plan")
    return False

  return True


while True:
  recharge_amount = int(input("Enter the recharge amount: "))

  if validate_recharge(recharge_amount):
    print("Recharge successful")
    break
  else:
    print("Recharge failed. Please try again.")




```

```python
#problem2: Inventory Reader Alert System

def check_inventory(products):
  for product in products:
    if products[product] < 15:
      print("Reorder Alert")

    else:
      print("Stock OK")

inventory = {
    "Laptop" : 20,
    "Mouse" : 10,
    "Keyboard" : 25,
    "Monitor" : 12

}

check_inventory(inventory)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
  base_fare = 50
  per_km = 12

  fare = base_fare + (distance * per_km)

  if peak == "yes":
    fare = fare + (fare *0.25)

  return fare

while True:
  distance = float(input("Enter distance in km : "))
  peak_hour = input("Is it peak hour (yes/no)")


  total_fare = calculate_fare(distance, peak_hour)
  print("Total fare $: ", total_fare)

  retry = input("Do you want to calculate another fare (yes/no)")

  if retry != "yes":
    break


```

```python
#Problem 5: Employee Attendance Eligibility Checker

def check_attendence(attendence_list):

  present_days = 0

  for day in attendence_list:
    if day == "P":
      present_days += 1

  attendance_percentage = (present_days / len(attendence_list)) * 100

  print("Attendance Percentage" , attendance_percentage)

  if attendance_percentage >= 75:
    return "Eligible"

  else:
    return "Not Eligible"

attendence = ['P', 'P', 'A', 'P', 'A', 'P','P', 'A', 'P']

status = check_attendence(attendence)

print("Check Status :" , status)

```

```python
# Problem 6: Password Strength Checker

def check_password(password):

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True

        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"


user_password = input("Enter password: ")

result = check_password(user_password)

print(result)
```
